```python
#Q2 Write a python program that will print the string of primes in a given range. For example - String of primes in a range of 0 - 1000, or 100 - 1000, etc.

def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def primes_in_range(start, end):
    prime_list = [num for num in range(start, end + 1) if is_prime(num)]
    return ", ".join(map(str, prime_list))

start_range = int(input("Enter the start of the range: "))
end_range = int(input("Enter the end of the range: "))

print("Prime numbers between", start_range, "and", end_range, "are:")
print(primes_in_range(start_range, end_range))

```

```python
# Q3  3) A coffee shop claims that their new coffee blend has significantly increased customer satisfaction compared to their previous blend. To test this claim, the coffee shop collects data on customer satisfaction for two different samples: one for the old blend and one for the new blend.
#Old Blend:

import numpy as np
from scipy.stats import t

# Sample data
mean_old = 3.8
std_old = 0.6
n_old = 25

mean_new = 4.2
std_new = 0.7
n_new = 30

# Calculate the test statistic
t_statistic = (mean_new - mean_old) / np.sqrt((std_new**2 / n_new) + (std_old**2 / n_old))

# Degrees of freedom
df = n_old + n_new - 2

# Calculate the critical value
alpha = 0.05
critical_value = t.ppf(1 - alpha / 2, df)

# Calculate the p-value
p_value = 2 * (1 - t.cdf(abs(t_statistic), df))

# Make a decision
if abs(t_statistic) > critical_value or p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

print("t-statistic:", t_statistic)
print("Critical value:", critical_value)
print("p-value:", p_value)




```

```python
#Q6 prompt:  In the online_retail.csv data, create a classification model that will classify the churning of the customers using any machine learning algorithm. We will use precision and recall for the evaluation of the model performance.

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score

# Load data
df = pd.read_csv("online_retail.csv")

# Data cleaning
df["Quantity"] = df["Quantity"].clip(lower=1)
df["UnitPrice"] = df["UnitPrice"].clip(lower=0.01)

# Feature engineering
df["Total_Spend"] = df["Quantity"] * df["UnitPrice"]
df["Customer_Age"] = 2023 - df["InvoiceDate"].dt.year

# Select features and target variable
X = df[["Total_Spend", "Customer_Age"]]
y = df["Churn"]

# Preprocess data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Model training
model = LogisticRegression()
model.fit(X_train, y_train)

# Model evaluation
y_pred = model.predict(X_test)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)


```
